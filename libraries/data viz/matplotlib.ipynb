# Matplotlib

Matplotlib is a library for making 2D plots in Python.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [8,4]
```

It is designed with the philosophy that you should be able to create simple plots with just a few commands.

```python
## 1. initialize

import matplotlib.pyplot as plt
import numpy as np

# 2. prepare data
x = np.linspace(0, 10*np.pi, 1000)
y = np.sin(x)

# 3. render
fig, ax = plt.subplots(figsize=(8,2))
ax.plot(x, y, color='blue')

# 4. configure axes
ax.set(xlim=(0,30), xticks=np.arange(0, 32, 2),
       ylim=(-1,1), yticks=np.arange(-1, 1.5, .5))

# 5. observe
plt.show()
```

Matplotlib offers several kinds of plots.

```python
fig, axs = plt.subplots(2,4, figsize=(14,5))

x = np.random.uniform(0, 1, 100)
y = np.random.uniform(0, 1, 100)

axs[0][0].scatter(x,y,s=1)

x = np.arange(10)
y = np.random.uniform(1, 10, 10)
axs[0][1].bar(x,y)

z = np.random.uniform(0, 1, (8,8))
axs[0][2].contourf(z)

z = np.random.uniform(0, 1, 4)
axs[0][3].pie(z)

z = np.random.normal(0, 1, 100)
axs[1][0].hist(z)

x = np.arange(5)
y = np.random.uniform(0, 1, 5)
axs[1][1].errorbar(x, y, y/4)

z = np.random.uniform(0, 1, (8,8))
axs[1][2].imshow(z)

z = np.random.normal(0, 1, (100,3))
axs[1][3].boxplot(z)

plt.show()
```

The anatomy of a plot.



Almost every element in a plot can be modified.

```python
fig, axs = plt.subplots(2,2,figsize=(12,4))

x = np.linspace(0,10,50)
y = np.sin(x)

axs[0][0].plot(x, y, color='black')
axs[0][1].plot(x, y, linestyle='--')
axs[1][0].plot(x, y, linewidth=2)
axs[1][1].plot(x, y, marker='o')

plt.show()
```

Labels are added to improve understanding of a plot.

```python
fig, ax = plt.subplots(figsize=(4,2))
ax.plot(x, y)

ax.set_title('A Sine Wave')
ax.set_ylabel('Y-axis')
ax.set_xlabel('X-axis')

plt.show()
```

Customize figure, axes, and spines.

```python
fig, axs = plt.subplots(2,2)
axs[0, 0].set_facecolor('#ddddff')
axs[1, 1].set_facecolor('#ffffdd')
```

```python
fig, ax = plt.subplots(figsize=(6,2))
ax.spines['top'].set_color('None')
ax.spines['right'].set_color('None')
```

Scatterplots can be enhanced by using transparency to show area with greater density.

```python
fig, ax = plt.subplots()
x = np.random.normal(-1, 1, 500)
y = np.random.normal(-1, 1, 500)

ax.scatter(x, y, 50, 'C1', lw=0, alpha=.3)
plt.show()
```
